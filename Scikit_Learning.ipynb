# Mastering scikit-learn's ColumnTransformer: Data Transformation Made Easy | ColumnTransformer

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/9 covid_toy.csv')
```

```python
df.head()
```

```python

df['gender'].value_counts()
```

```python
df['city'].value_counts()
```

```python

from sklearn.model_selection import train_test_split
X = df.drop(columns=['has_covid'])
Y = df['has_covid']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train.shape,Y_train.shape,Y_test.shape,X_test.shape
```

```python
X_train
```

**Without column Transformer**

```python
# missing value fill
from sklearn.impute import SimpleImputer
si = SimpleImputer()#default='mean'

X_train_fever = si.fit_transform(X_train[['fever']])
X_test_fever = si.fit_transform(X_test[['fever']])
X_train_fever.shape
```

```python
X_train_fever_df = pd.DataFrame(X_train_fever, columns=['fever'])
display(X_train_fever_df.head())
```

```python
# encoding categorical value
```

```python
X_train_fever_df.isnull().sum()
```

```python
# encoding categorical values
from sklearn.preprocessing import OrdinalEncoder
Ordinal = OrdinalEncoder(categories=[['Mild','Strong']])
X_train_cough = Ordinal.fit_transform(X_train[['cough']])
X_test_cough = Ordinal.fit_transform(X_test[['cough']])
X_train_cough.shape
```

```python
# gender and city
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder(drop='first',sparse_output=False)
X_train_gender_city = OHE.fit_transform(X_train[['gender','city']])
X_test_gender_city = OHE.fit_transform(X_test[['gender','city']])
X_train_gender_city.shape
```

```python
# Extracting Age
X_train_age = X_train.drop(columns=['gender','fever','cough','city']).values
X_test_age = X_test.drop(columns=['gender','fever','cough','city']).values
X_train_age.shape
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_gender_city,X_train_cough), axis=1)
X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough), axis=1)
```

```python
# Get the feature names for the one-hot encoded columns
ohe_cols = OHE.get_feature_names_out(['gender', 'city'])

# Create a list of all column names in the correct order
all_cols = ['age', 'fever'] + list(ohe_cols) + ['cough']

# Convert the transformed data to a DataFrame
X_train_transformed_df = pd.DataFrame(X_train_transformed, columns=all_cols)

# Display the head of the new DataFrame
display(X_train_transformed_df.head())
```

# Function Transformer In Sklearn | Feature Transformations in Machine Learning

We used this function if our data is not normally disrtibuted

```python
import numpy as np
import pandas as pd

import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import FunctionTransformer
```

```python
df = pd.read_csv('/content/10 titanic.csv',usecols=['Age','Fare','Survived'])
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Fare'].fillna(df['Fare'].mean(),inplace=True)
```

```python
df
```

```python
X = df.iloc[:,1:3]
Y = df.iloc[:,0]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
fig, axs = plt.subplots(2,2,figsize=[14,8])

# plot the first distribution plot for age
axs[0,0].hist(X_train['Age'],bins=20, density=True,alpha=0.6,color='g',edgecolor='black')
axs[0,0].set_title('Distribution Plot - Age')

# plot the first QQ plot for Age
stats.probplot(X_train['Age'],dist='norm',plot=axs[0,1])
axs[0,1].set_title('Distribution plot-Fare')

# plot the second distribution plot for Fare
axs[1,0].hist(X_train['Fare'],bins=20,density=True,alpha=0.6,color='b',edgecolor='black')
axs[1,0].set_title('Distribution Plot - Fare')

# Plot the second QQ plot for Fare
stats.probplot(X_train['Fare'],dist='norm',plot=axs[1,1])
axs[1,1].set_title('QQ Plot - Fare')

plt.tight_layout()# Adjust to prevent overlapping
plt.show()

```

**Training models before function transformer**

```python
clf = LogisticRegression()
clf2 = DecisionTreeClassifier()
clf.fit(X_train,Y_train)
clf2.fit(X_train,Y_train)
Y_pred = clf.predict(X_test)
Y_pred1 = clf2.predict(X_test)
```

```python
print("Accuracy LR",accuracy_score(Y_test,Y_pred))
print("Accuracy DT",accuracy_score(Y_test,Y_pred1))
```

**Log Transformer**

```python
trf = FunctionTransformer(func=np.log1p)
X_train_transformed = trf.fit_transform(X_train)
X_test_transformed = trf.transform(X_test)
```

```python
X_train,X_train_transformed
```

```python
clf.fit(X_train_transformed,Y_train)
clf2.fit(X_train_transformed,Y_train)
Y_pred = clf.predict(X_test_transformed)
Y_pred1 = clf2.predict(X_test_transformed)

print("Accuracy LR",accuracy_score(Y_test,Y_pred))
print("Accuracy DT",accuracy_score(Y_test,Y_pred1))
```

```python
fig, axs = plt.subplots(2,2,figsize=[14,8])

# plot the first distribution plot for age
stats.probplot(X_train['Age'],dist='norm',plot=axs[0,0])
axs[0,0].set_title('Age before Log')

# plot the first QQ plot for Age
stats.probplot(X_train_transformed['Age'],dist='norm',plot=axs[0,1])
axs[0,1].set_title('Age after Log')

# plot the second distribution plot for Fare
stats.probplot(X_train['Fare'],dist='norm',plot=axs[1,0])
axs[1,0].set_title('Fare Before Log')

# Plot the second QQ plot for Fare
stats.probplot(X_train_transformed['Fare'],dist='norm',plot=axs[1,1])
axs[1,1].set_title('Fare After Log')

plt.tight_layout()# Adjust to prevent overlapping
plt.show()
```
