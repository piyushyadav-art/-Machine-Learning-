```python
# importing the dependencies
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

We will be Working on the breast cancer dataset

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
data_frame['label'].value_counts()
```

1 -->Benign

2 -->Malignant



Separating the feature and target

```python
X = data_frame.drop(columns='label',axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(X)
```

```python
print(Y)
```

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

#**Gride search CV**

It is used to determining the best parameterfor our model

```python
# Loading the SVC model
model = SVC()
```

```python
# hyperparameter

parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# grid search
classifier = GridSearchCV(model,parameters,cv=5)
```

```python
# fitting the data to our model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# best parameters
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# loading the result to a pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

Highest Accuracy = 95.2 %,

Best parameter={'C:10','Kernal':'linear'}


#**RandomizedSearchCV**

```python
# Loading the SVC model
model = SVC()
```

```python
# hyperparameter

parameters = {
    'kernel':['linear','poly','rbf','sigmoid'],
    'C':[1,5,10,20]
}
```

```python
# grid search
classifier = RandomizedSearchCV(model,parameters,cv=5)
```

```python
# fitting the data to our model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# best parameters
best_parameters = classifier.best_params_
print(best_parameters)
```

```python
# highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# loading the result to a pandas dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
randomized_search_result
```

Highest Accuracy = 95.2 %,

Best parameter={'C:10','Kernal':'linear'}
