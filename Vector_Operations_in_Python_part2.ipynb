```python
import numpy as np
```

**Dot Product of two vectors**

```python
a = np.array([2,3])
b = np.array([4,4])

a_dot_b = np.dot(a,b)
print(a_dot_b)
```

```python
c = np.array([40,50,35])
d = np.array([53,24,68])

c_dot_d = np.dot(c,d)
print(c_dot_d)
```

**Cross Product of two vector**

```python
a = np.array([2,3])
b = np.array([4,4])

a_cross_b = np.cross(a,b)

print(a_cross_b)
```

```python
c = np.array([5,10,20])
d = np.array([18,32,50])

c_cross_d = np.cross(c,d)
print(c_cross_d)
```

**Screenshot 2025-08-07 120514.pngProjection of vector "a" on vector "b"**

```python
a = np.array([2,5])
b = np.array([8,-6])

# magnitude of "b"vector
magnitude_of_b = np.sqrt(sum(b**2))

proj_of_a_b = (np.dot(a,b)/magnitude_of_b**2)*b

print("Projection of a vector on 'b' vector",proj_of_a_b)
```
