# Bag of Words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'text':['people watch campusx','campusx watch campusx','people write comment','campusx write comment'],'output':[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
# vocab
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

```python
cv.transform(["campusx watch and write comment of campusx"]).toarray()
```

Removal Punctuation


```python
import string
import re
```

```python
PUNCTUATION_PATTERN = f'[{re.escape(string.punctuation)}]'
```

```python
df['review'] = (
    df['review']
    .astype(str)
    .str.replace(
        PUNCTUATION_PATTERN,
        '',  # Replace the matched pattern with an empty string
        regex=True,
        flags=re.IGNORECASE
    )
)
```

Removing Stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
  new_text = []
  for word in text.split(): # Split the text into words
    if word.lower() not in stopwords.words('english'): # Check for lowercase word in stopwords
      new_text.append(word)
  return " ".join(new_text)
```

```python
import gensim
import pandas as pd
```

```python
df = pd.read_json('/content/reviews_Cell_Phones_and_Accessories_5.json.gz', lines=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.reviewText[0]
```

```python
gensim.utils.simple_preprocess(df.reviewText[0])
```

```python
review_text = df.reviewText.apply(gensim.utils.simple_preprocess)
review_text
```

```python
Model = gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4,
)
```

```python
Model.build_vocab(review_text, progress_per=1000)
```

```python
Model.epochs
```

```python
Model.train(review_text, total_examples=Model.corpus_count, epochs=Model.epochs)
```

```python
Model.save('/content/word2vec-amazon-cell-accessories-reviews-short.model')
```

```python
Model.wv.most_similar("bad")
```

```python
Model.wv.similarity(w1="cheap", w2="inexpensive")
```

```python
Model.wv.similarity(w1="good", w2="wasted")
```

```python
Model.wv.similarity(w1="good", w2="great")
```

```python
Model.wv.similarity(w1="great", w2="product")
```
