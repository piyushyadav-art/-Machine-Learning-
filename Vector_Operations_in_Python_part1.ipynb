```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Ploting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-4,-8,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vector

```python
vector_1=np.array([0,0,2,3])
vector_2=np.array([0,0,3,-2])

sum = vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of 2 vector

```python
vector_1=np.array([0,0,2,3])
vector_2=np.array([0,0,3,-2])

sub = vector_1 - vector_2
print(sub)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by a scaler

```python
vector_1 =np.array([0,0,2,3])
vector_2 = 2
mult = vector_1*vector_2
print(mult)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,4,6,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 =np.array([0,0,2,3])
vector_2 = -0.5
mult = vector_1*vector_2
print(mult)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-1,-1.5,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
