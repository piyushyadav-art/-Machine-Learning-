Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/images?q=tbn:ANd9GcRErDuS2HPi8wlmIEJaewbgJZ8fFIJCyxBZYQ&s')
```

```python
# Resize the image to 200x200 pixels using the BILINEAR resampling filter
img_resized = img.resize((200, 200), Image.Resampling.BILINEAR)
```

```python
print(img_resized)
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
# displaying the image from numpy array
img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plt = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Convert the Rgb images to Grayscale image using OpenCV

```python
# importing OpenCV library
import cv2
```

```python
img = cv2.imread('/content/images?q=tbn:ANd9GcRErDuS2HPi8wlmIEJaewbgJZ8fFIJCyxBZYQ&s')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python

type(grayscale_image)
```
